```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x**2
y
```

```python
plt.plot(x, y, 'g')
plt.xlabel('X Axis label')
plt.ylabel('Y Axis label')
plt.title('This is the Title')
plt.show()
```

Creating Multiplots on Same Canvas

```python
plt.subplot(1 , 2 , 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b:')
```

OO APROACH

```python
#Add Canvas Figure
fig = plt.figure()

#Add set of axes to figure
axes = fig.add_axes([0.1 , 0.1 , 0.9 , 0.9]) #list ->all quantities =fractions of figure

#Plot and set labels
axes.plot(x, y, 'r-')
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('This is the Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1 ,0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

axes1.plot(x, y, 'r')
axes2.plot(y, x, 'b')
```

```python
axes1.plot(x, y, 'r')
axes2.plot(y, x, 'b')
```

```python
#Legends
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label='x**2' )
ax.plot(x, x**3, label='x**3')

ax.legend()
```

```python
#Legends
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label='x**2' )
ax.plot(x, x**3, label='x**3')

ax.legend(loc = 1) 
#loc=1 upper right corner, loc=2 upper left corner, loc=3 lower left, loc=4 lower right corner
```
